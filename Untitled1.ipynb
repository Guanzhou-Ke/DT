```python
# load all data
df = pd.read_csv("C:/Users/xiaol/PycharmProjects/zzw/Data/data.csv")
```

```python
# remove some features
# 'Adensity','AC_material_weight','TD_weight','Measure_Velocity','Overpressure'
df = df.drop(columns=['Overpressure'])
```

```python
# normalize function
def normalize(column):
    return ((column - column.mean()) / column.std())

data_norm = df.iloc[:,:-1].apply(normalize)
data_norm
```

```python
data_norm['React_Efficiency'] = df['React_Efficiency']
data_norm
```

```python
# heatmap
plt.figure(figsize=(16, 5))
c = data_norm.corr()
sns.heatmap(c, cmap="BrBG", annot=True)
c
```

```python
# construct train_data and test_data 
# way 1 : select 4 group Zr&Ti as train data

test_data = pd.concat([data_norm.iloc[27:34], data_norm.iloc[59:65]])
select_test_indices = list(range(27, 34)) + list(range(59, 65))
train_data = data_norm.drop(select_test_indices)

X_train = train_data.drop(columns=['React_Efficiency'])
y_train = train_data['React_Efficiency']
X_test = test_data.drop(columns=['React_Efficiency'])
y_test = test_data['React_Efficiency']
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = sum((y_pred - y_test)**2)
print('mse:', mse)
print('y_test:', y_test)
print('y_pred:', y_pred)
```

```python
# MLP
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
```

```python
X = X_train.values
y = y_train.values
X_train = torch.tensor(X, dtype=torch.float32)
y_train = torch.tensor(y, dtype=torch.float32).view(-1, 1)
```

```python
X_test, y_test = X_test.values, y_test.values
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)
```

```python
class Net(nn.Module):
    def __init__(self, input_dim):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 1)  # 输出层，输出一个值，因为是回归任务

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
# 初始化模型
input_dim = X_train.shape[1]
model = Net(input_dim)

# 定义损失函数和优化器
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.005)

# 训练模型
num_epochs = 70
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# 评估模型
model.eval()
with torch.no_grad():
    y_pred = model(X_test)
    mse = criterion(y_pred, y_test)
    mse_sum = torch.sum((y_pred - y_test)**2)
    print(mse_sum.item())
    print("Mean Squared Error:", mse.item())
```

```python
print('nn y_pred:', y_pred)
print('nn y_test:', y_test)

tmp = torch.arange(len(y_pred))
plt.plot(tmp, y_pred, label='y_pred')
plt.plot(tmp, y_test, label='y_test')
plt.legend()
plt.show()
```
